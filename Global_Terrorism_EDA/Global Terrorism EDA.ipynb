# Extracting columns:

```python
cols = ['iyear', 'imonth', 'iday', 'country_txt', 'city', 'latitude', 'longitude', 'location', 'attacktype1_txt', 'targettype1_txt', 'targsubtype1_txt', 'target1', 'gname', 'motive', 'weaptype1_txt', 'dbsource', 'region_txt', 'nkill', 'nwound', 'natlty1_txt', 'weapdetail']
```

```python
terror = pd.DataFrame(data=df, columns=cols)
```

# Data PreProcessing:

```python
terror.shape
```

```python
terror.isnull().sum()
```

```python
terror.head()
```

```python
terror.rename(columns={'iyear':'year',
                       'imonth':'month',
                       'iday':'day',
                       'country_txt':'country_name',
                       'attacktype1_txt':'attacktype',
                       'targettype1_txt':'targtype',
                       'gname':'group_name',
                       'dbsource':'source',
                       'region_txt':'region1',
                       'nkill':'killed',
                       'nwound':'wounded',
                       'weapdetail':'weapon_detail',
                       'natlty1_txt':'nationality'}, inplace=True)
```

```python
terror.head(1)
```

```python
pd.set_option("display.max_columns", 500)
```

```python
terror.head()
```

```python
terror['casualities'] = terror['wounded'] + terror['killed']
```

```python
terror.head(1)
```

```python
nationality_top = terror[terror['nationality'] != 'Unknown']
```

```python
nationality_type = nationality_top['nationality'].value_counts().reset_index()
```

```python
nationality_type.rename(columns={'Counts':'Nationality',
                                 'count':'Counts'}, inplace=True)
```

```python
nationality_type
```

```python
terror['nationality'].nunique()
```

```python
terror['nationality'].unique()
```

```python
terrorism = terror[terror['group_name'] != 'Unknown']
```

```python
Terrorism_group_org = terrorism['group_name'].value_counts().reset_index()
```

```python
Terrorism_group_org.rename(columns={'Counts':'Group_name',
                                    'count':'Counts'}, inplace=True)
```

```python
Terrorism_group_org
```

```python
terror.head()
```

```python
city1 = terror[terror['city'] != 'Unknown']
```

```python
city_cln = city1['city'].value_counts().reset_index()
```

```python
city_cln.rename(columns={'Counts':'City_name',
                         'count':'Counts'}, inplace=True)
```

```python
city_cln
```

```python
# Weapon type:
wp = terror[terror['weaptype1_txt'] != 'Unknown']
weaptype_cln = wp['weaptype1_txt'].value_counts().reset_index()
weaptype_cln.rename(columns={'weaptype1_txt':'weapon_type',
                             'count':'Counts'}, inplace=True)
```

```python
weaptype_cln
```

# Exploratory Data Analysis (EDA):

```python
terror.head(1)
```

```python
# NATIONALITY:
f, ax = plt.subplots(figsize = (10,5))
ax =  sns.barplot(x = 'Counts', y = 'nationality', data=nationality_type[:10],
                  palette = 'rocket').set_title('NATIONS SUFFERED MOST OF THE TARGET ATTACKS')
```

```python
# Groups:
f, ax = plt.subplots(figsize = (10,5))
ax = sns.barplot(x='Counts', y='group_name', data=Terrorism_group_org[:10],
                 palette='mako')
ax.set_title('GROUPS RESPONSIBLE FOR THE MOST ATTACKS')
ax.set_ylabel('Terrorism Groups')
```

```python
# City:
f, ax = plt.subplots(figsize = (10,5))
ax = sns.barplot(x='Counts', y='city', data=city_cln[:15],
                 palette='cubehelix')
ax.set_title('TOP 15 CITIES SUFFERED MOST TERRORISM ATTACKS')
ax.set_ylabel('Cities')
```

```python
# Weapon type:
f, ax = plt.subplots(figsize = (10,5))
ax = sns.barplot(x='Counts', y='weapon_type', data=weaptype_cln[:5])
ax.set_title('WEAPONS USED FOR ATTACKS')
ax.set_ylabel('Weapons')
```

```python
# Attack type:
plt.subplots(figsize = (12,8), dpi=100)
sns.countplot(y='attacktype', data=terror, palette='coolwarm', order=terror['attacktype'].value_counts().index)
plt.yticks(size = 15)
plt.xticks(size = 15)
plt.xlabel('Counts', size=15)
plt.ylabel('Attack types', size=15)
plt.title('ATTACK TYPES BY TERRORISTS', size = 20)
```

```python
# Target type:
plt.subplots(figsize = (15,7), dpi=100)
sns.countplot(x='targsubtype1_txt', data=terror, order=terror['targsubtype1_txt'].value_counts().index[:25])
plt.xticks(rotation = 90)
plt.xlabel('Target types', size=15)
plt.ylabel('Count', size=15)
plt.title('TOP 25 TARGET TYPES')
```

```python
plt.figure(figsize = (10,25), dpi=120)
sns.histplot(data=terror, y='targsubtype1_txt', bins=15, kde=True)
plt.xlabel('Order counts')
plt.title('TARGET TYPE ATTACKED')
```
